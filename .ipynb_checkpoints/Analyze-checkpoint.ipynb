## Import Library and data


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, log_loss
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler ,LabelEncoder 
import warnings
warnings.filterwarnings('ignore')

# read
inputs = pd.read_csv('outputs2.csv', encoding="utf-8")

inputs.head()
```

## Dataset Distribution

```python
import seaborn as sns
# transform safty_flag to int data type, and store it to a new attribute called safetly_flag_numerical
# relationship between safety_flag and safetly_flag_numerical are: 'False' => 0， 'True' => 1

inputs['safety_flag_numerical'] = inputs['safety_flag'].map(lambda x: 1 if x else 0)
# get the numercial attributt columns


numerical = ['confidencescore', 'estimatedvalue', 'fsd', 'safety_flag_numerical']

# plot the graph
g = sns.pairplot(inputs[numerical], 
                 hue='safety_flag_numerical', 
                 palette='seismic', 
                 diag_kind = 'kde',
                 diag_kws=dict(shade=True))

g.set(xticklabels=[])
```

## Create Model
    Data train tools from scikit-learn
    - SVM
    - DecisionTree
    - GaussianNB
    - K Nearest Neighbors
    - LogisticRegression
       

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

X = inputs[['confidencescore', 'estimatedvalue', 'fsd']]
Y = inputs['safety_flag_numerical']

# Transforms features by scaling each feature to a 0-1.
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# split the dataset into 80% of training data, 20% of the testing data with train_test_split 
x_train, x_test, y_train, y_test = train_test_split(X, Y,  test_size=1/4, random_state = 0)

```

### SVM


```python
from sklearn.svm import SVC

# c value accurate
svm_acc=[0.001, 0.01, 0.1, 1, 10, 100]

for v in svm_acc:
    svm_model = SVC(C=v)
    svm_model.fit(x_train, y_train)
    acc = svm_model.score(x_test, y_test)
    print('c=%s, accurate: %.3f' % (v, acc))
```

### DecisionTree


```python
from sklearn.tree import DecisionTreeClassifier

# Depth of the tree
depths=[1,3,5,7,9]


for v in depths:
    dt_model = DecisionTreeClassifier(max_depth=v)
    dt_model.fit(x_train, y_train)
    acc = dt_model.score(x_test, y_test)
    print('depths=%s, accurate: %.3f' % (v, acc))

    
```

### GaussianNB


```python
from sklearn.naive_bayes import GaussianNB

gnb_model = GaussianNB()
gnb_model.fit(x_train, y_train)
acc = gnb_model.score(x_test, y_test)
print('accurate: %.3f' % acc)
```

### K Nearest Neighbors


```python
from sklearn.neighbors import KNeighborsClassifier

neighbors = [1,3,5,7,9]

for v in neighbors:
    knn = KNeighborsClassifier(n_neighbors=v)
    knn.fit(x_train, y_train)
    acc = knn.score(x_test, y_test)
    print('accurate: %.3f' % acc)
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression


lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
acc = lr_model.score(x_test, y_test)
print('accurate: %.3f' % acc)
```
